```python
'''
Program: Group 5 Code.ipynb
Group: 5 - Python Pioneers 9000
Authors: Tim Guichard, Uma Maheshwari,
         Sakeet Kopparapu, Anil Kumar Vonti Konda
Date: 11/20/2024

This program contains the code from each member of the group compiled together
into a single program. Each member's code is contained within it's own function.
Two additional functions support data import, cleanup, and creation of a sub frame.
'''
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


def dataImport(path):
    '''Requires a path argument to data set. Imports the data set and converts to data frame.
    Drops unneeded columns, and converts relevant columns to numerical data types'''
    # Original
    frame = pd.read_csv(path)

    # Working Copy
    workingFrame = frame.copy()

    # Drop unwanted columns
    workingFrame.drop(columns=["ISRC", "TIDAL Popularity"], inplace= True)

    # Change columns with numbers into number data types
    # Keep these as string/objects
    stringColumns = ["Track", "Album Name", "Artist", "Release Date"]

    workingFrame.fillna(value= 0, inplace= True)

    for column in workingFrame.columns:
        if column not in stringColumns and workingFrame[column].dtype == "object":
            workingFrame[column] = workingFrame[column].str.replace(",", "")
            workingFrame[column].fillna(value= 0, inplace= True) # Second fillna needed for unknown reason
            workingFrame[column] = pd.to_numeric(workingFrame[column], downcast="integer")


    workingFrame.dtypes

    return workingFrame


def createDataSubset(frame):
    """Requires a data frame argument. Uses a series of substring checks to eliminate unwanted columns.
    Returns a new data frame containing only select columns"""

    # Substring of column names to be kept
    keepStrings = ["Track", "Streams", "Views"]

    # Substrings present in kept column names to be removed
    removeStrings = ["Score", "Stations", "Explicit"]

    keepList = []
    removeList = []

    # First filter
    for columnName in frame.columns:
        for label in keepStrings:
            if label in columnName:
                keepList.append(columnName)
                break
    print("First filter - Keeping the following columns:")
    print(keepList)

    # Second filter
    for index, column in enumerate(keepList):
        for label in removeStrings:
            if label in column:
                keepList.pop(index)

    print("Second Filter - Keeping the following columns:")
    print(keepList)

    for column in frame.columns:
        if column not in keepList:
            removeList.append(column)

    print("Columns to be removed:")
    print(removeList)

    return frame.drop(columns= removeList)


def timCode():
    '''Tim's code. Contains creation of a calculated column, and calulations and visualizations for
    top tracks by streams/views and explicit vs clean song comparisons.'''

    # Create calculated column, add to data frame, and sort

    # Remove commas in columns, set NA to 0
    streamFrame.fillna(value= 0, inplace= True)

    streamFrame["Total Streams/Views"] = (
        streamFrame["Spotify Streams"] +
        streamFrame["YouTube Views"] +
        streamFrame["TikTok Views"] +
        streamFrame["Pandora Streams"] +
        streamFrame["Soundcloud Streams"]
    )

    # Sort by Total column descending
    streamFrame.sort_values(by=["Total Streams/Views"],ascending=False, inplace=True)

    print(streamFrame.head(10))

    # Top 10 Songs - Total Streams and Views
    topFrame = streamFrame.head(10)

    tracks = list(topFrame["Track"])
    totals = list(topFrame["Total Streams/Views"])


    plt.figure(figsize = (10, 8))
    plt.bar(tracks, totals)

    plt.xticks(rotation=45, ha="right")
    plt.yticks([0, 50000000000, 100000000000, 150000000000, 200000000000],["0", "50 Billion", "100 Billion", "150 Billion", "200 Billion"])

    plt.title("Top Ten Streamed/Viewed Songs - All Platforms")
    plt.xlabel("Track Name")
    plt.ylabel("Number of Streams/Views")

    plt.show()

    # Top 10 songs with the two highest outliers removed
    topFrame2 = streamFrame.copy()

    # Drop the top two by index labels
    topFrame2 = streamFrame.drop(index=[1024, 1922])
    topFrame2 = topFrame2.head(10)

    print(topFrame2)

    # Create graph with removed outliers

    tracks = list(topFrame2["Track"])
    totals = list(topFrame2["Total Streams/Views"])

    plt.figure(figsize = (10, 8))
    plt.bar(tracks, totals)

    plt.xticks(rotation=45, ha="right")
    plt.yticks([0, 1*10**10, 2*10**10, 3*10**10, 4*10**10, 5*10**10, 6*10**10],
            [0, 10, 20, 30, 40, 50, 60])

    plt.title("Top Ten Streamed/Viewed Songs - All Platforms - Outliers Removed")
    plt.xlabel("Track Name")
    plt.ylabel("Number of Streams/Views (in billions)")

    plt.show()


    # Explicit vs Clean songs
    # Utilizes Explicit Track Boolean Flag (int)

    # Get number of clean and explicit tracks
    cleanTracks = len(workingFrame[workingFrame["Explicit Track"] == 0])
    explicitTracks = len(workingFrame[workingFrame["Explicit Track"] == 1])

    # Validate Data

    print("Total Tracks:", len(workingFrame.index))
    print("Explicit Tracks:", explicitTracks)
    print("Clean Tracks:", cleanTracks)

    # Create Pie Chart

    plt.figure(figsize=[10, 8])
    plt.pie([explicitTracks, cleanTracks], labels=["Explicit", "Clean"],
            colors= ["red", "blue"], autopct="%1.1f%%",
            textprops={'size': 'larger'})

    plt.title("Ratio of Explicit to Clean Tracks")

    plt.show()


    # Scatter plot

    # Sort by index to add back Explicit Track column
    unsortedStreamFrame = streamFrame.sort_index()
    unsortedStreamFrame["Explicit Track"] = workingFrame["Explicit Track"]

    # Validate
    print(unsortedStreamFrame.head())

    tracks = list(unsortedStreamFrame.index)
    streamCount = list(unsortedStreamFrame["Total Streams/Views"])

    # Build color list to correspond with Explicit Track flag
    colors = []
    for flag in unsortedStreamFrame["Explicit Track"]:
        if flag == 1:
            colors.append("red")
        else:
            colors.append("blue")

    plt.figure(figsize= [20, 15])

    # Remove record indexes from x axis
    plt.xticks([])
    # Use log scale instead of linear to get a better visual
    plt.gca().set_yscale('log')

    # Actual plot
    plt.scatter(tracks, streamCount, c=colors)
    # Empty subplot to force legend to populate correctly with explicit tracks
    plt.scatter([], [], c="red")


    plt.ylabel("Total Streams/Views (logarithmic scale)", fontsize= 15)
    plt.title("Distribution of Total Streams/Views by Track and Explicit Rating", fontsize= 15)
    plt.legend(["Clean Track", "Explicit Track"], loc= "upper left", fontsize= 15, markerscale=2.0)
    plt.show()


def sakeetCode():
    '''Sakeet's code. Contain's calulations and visualizations for top artists by
    number of songs on the list and total streams/views'''


    # Top Artist by number of songs in data set
    workingFrame['Track'] = workingFrame['Track'].astype('category').cat.codes
    workingFrame['Spotify Streams'] = workingFrame['Spotify Streams'].astype('category').cat.codes
    workingFrame['YouTube Views'] = workingFrame['YouTube Views'].astype('category').cat.codes
    workingFrame['TikTok Views'] = workingFrame['TikTok Views'].astype('category').cat.codes
    workingFrame['Pandora Streams'] = workingFrame['Pandora Streams'].astype('category').cat.codes
    workingFrame['Soundcloud Streams'] = workingFrame['Soundcloud Streams'].astype('category').cat.codes

    X = workingFrame.loc[:,['YouTube Views', 'Spotify Streams', 'TikTok Views', 'Pandora Streams', 'Soundcloud Streams']]
    Y= workingFrame.loc[:, workingFrame.columns == 'Track' ]

    categorical_var = ['Track', 'YouTube Views', 'Spotify Streams', 'TikTok Views', 'Pandora Streams', 'Soundcloud Streams']
    workingFrame[categorical_var] = workingFrame[categorical_var].astype('category')

    top_artist = workingFrame['Artist'].value_counts().idxmax()
    top_artist_count = workingFrame['Artist'].value_counts().max()

    print(f"The artist who appears the most is '{top_artist}', with {top_artist_count} appearances.")

    workingFrame.groupby("Artist")['Track'].agg("count").sort_values(ascending=False).head(5)
    workingFrame.groupby("Artist")['Album Name'].agg("count").sort_values(ascending=False).head(5)
    workingFrame.groupby("Artist")['Release Date'].agg("count").sort_values(ascending=False).head(5)

    stream_columns = [
    'Spotify Streams',
    'YouTube Views',
    'TikTok Views',
    'Pandora Streams',
    'Soundcloud Streams'
    ]

    # Convert relevant columns to numeric (handling commas in numbers)
    for col in stream_columns:
        workingFrame[col] = workingFrame[col].replace(',', '', regex=True).astype(float)

    # Calculate the total streams/views for each artist
    workingFrame['Total Streams/Views'] = workingFrame[stream_columns].sum(axis=1)
    top_artists = workingFrame.groupby('Artist')['Total Streams/Views'].sum().sort_values(ascending=False).head(10)

    # Display the top artists
    print("Top 10 Artists by Total Streams/Views:")
    print(top_artists)

    artist_song_count = workingFrame['Artist'].value_counts()

    # Plot the data
    plt.figure(figsize=(10, 6))
    artist_song_count[:10].plot(kind='bar', color='skyblue')  # Display top 10 artists
    plt.title('Top 10 Artists by Number of Songs', fontsize=16)
    plt.xlabel('Artist', fontsize=14)
    plt.ylabel('Number of Songs', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    # Show the chart
    plt.show()


    # Plot the data
    plt.figure(figsize=(12, 7))
    top_artists.plot(kind='bar', color='purple', alpha=0.7)
    plt.title('Top 10 Artists by Total Streams/Views', fontsize=16)
    plt.xlabel('Artist', fontsize=14)
    plt.ylabel('Total Streams/Views (in billions)', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    # Show the chart
    plt.show()


def umaCode():
    '''Uma's code. Contain's calulations and visualizations for song release dates'''

    # Read the CSV file
    df = workingFrame.copy()

    # Convert Release Date to datetime
    df['Release Date'] = pd.to_datetime(df['Release Date'])
    # Extract year from Release Date
    df['Year'] = df['Release Date'].dt.year


    # Create figure with two y-axes
    fig, ax1 = plt.subplots(figsize=(15, 7))

    # Plot song count
    sns.histplot(data=df, x='Year', bins=30, color='darkblue', alpha=0.5, ax=ax1)
    ax1.set_xlabel('Release Year')
    ax1.set_ylabel('Number of Songs', color='darkblue')
    ax1.tick_params(axis='y', labelcolor='darkblue')

    # Create second y-axis for average streams
    ax2 = ax1.twinx()
    year_avg_streams = df.groupby('Year')['Spotify Streams'].mean() / 1_000_000  # Convert to millions
    ax2.plot(year_avg_streams.index, year_avg_streams.values, color='red', linewidth=2)
    ax2.set_ylabel('Average Streams (Millions)', color='red')
    ax2.tick_params(axis='y', labelcolor='red')

    plt.title('Song Frequency vs Average Streams by Year')
    plt.show()

    # Calculate some key statistics
    recent_songs = df[df['Year'] >= 2020]
    older_songs = df[df['Year'] < 2020]

    print("\
    Key Statistics:")
    print(f"Average streams for songs 2020-2024: {recent_songs['Spotify Streams'].mean()/1_000_000:.2f} million")
    print(f"Average streams for songs before 2020: {older_songs['Spotify Streams'].mean()/1_000_000:.2f} million")
    print(f"\
    Number of songs 2020-2024: {len(recent_songs)}")
    print(f"Number of songs before 2020: {len(older_songs)}")

    # Extract month from Release Date
    df['Month'] = df['Release Date'].dt.month

    # Count the number of releases for each month
    monthly_counts = df['Month'].value_counts().sort_index()

    print("Monthly release counts:")
    print(monthly_counts)

    # Plot the number of releases per month
    plt.figure(figsize=(10, 6))
    sns.barplot(x=monthly_counts.index, y=monthly_counts.values, palette='viridis')
    plt.title('Number of Music Releases by Month')
    plt.xlabel('Month')
    plt.ylabel('Number of Releases')
    plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    plt.show()


def anilCode():
    '''Anil's code. Contain's calculations and visualizations for the most popular platforms
    and top songs on each platform'''


    # Reimport and clean due to changes being made by previous functions
    workingFrame = dataImport("/content/Most Streamed Spotify Songs 2024 - UTF-8.csv")

    # Define columns related to different platforms
    platform_columns = [
        'Spotify Streams', 'YouTube Views', 'TikTok Views',
        'Apple Music Playlist Count', 'Amazon Playlist Count',
        'Pandora Streams', 'Soundcloud Streams'
    ]

    # Clean and convert columns to numeric where needed
    for col in platform_columns:
        workingFrame[col] = pd.to_numeric(workingFrame[col].replace({',': ''}, regex=True), errors='coerce')

    # Calculate the mean for each platform column
    platform_stats = workingFrame[platform_columns].agg(['mean'])

    # Identify the most popular platform based on the highest mean
    most_popular_platform = platform_stats.idxmax(axis=1)['mean']
    highest_mean_value = platform_stats.max(axis=1)['mean']

    # Display the result with a final conclusion
    print(platform_stats)
    print(f"\nThe most popular platform based on the average (mean) values is: {most_popular_platform}, with an average count of {highest_mean_value:.2f}.")

    # Data from your output
    platforms = ['Spotify Streams', 'YouTube Views', 'TikTok Views',
                'Apple Music Playlist Count', 'Amazon Playlist Count',
                'Pandora Streams', 'Soundcloud Streams']
    means = [4.363971e+08, 3.758269e+08, 9.135283e+08,
            47.943913, 19.535217, 6.506994e+07, 4.089647e+06]

    # Create a pie chart
    plt.figure(figsize=(10, 7))
    wedges, texts, autotexts = plt.pie(means, labels=platforms, autopct='%1.1f%%', startangle=140, labeldistance=1.2)

    # Modify the text properties to prevent overlap
    for text in texts:
        text.set_fontsize(10)  # You can adjust this value for readability

    plt.gca().set_aspect('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title('Most Popular Platforms')

    # Save the figure with extra space
    plt.savefig('pie_chart.png', bbox_inches='tight', pad_inches=0.5)

    # Show the plot
    plt.show()

    # Define columns related to different platforms
    platform_columns = [
        'Spotify Streams', 'YouTube Views', 'TikTok Views',
        'Apple Music Playlist Count', 'Amazon Playlist Count',
        'Pandora Streams', 'Soundcloud Streams'
    ]

    # Dictionary to store top 3 songs for each platform
    top_songs_per_platform = {}

    # For each platform, get the top 3 unique songs based on the highest values in each column
    for platform in platform_columns:
        # Convert the column to numeric in case of non-numeric values
        workingFrame[platform] = pd.to_numeric(workingFrame[platform].replace({',': ''}, regex=True), errors='coerce')

        # Get top 3 unique songs by dropping duplicates in the 'Track' column
        top_songs_per_platform[platform] = workingFrame.drop_duplicates(subset='Track').nlargest(3, platform)[['Track', platform]]

    # Display the top 3 unique songs for each platform
    for platform, top_songs in top_songs_per_platform.items():
        print(f"Top 3 songs on {platform}:")
        print(top_songs)
        print("\n")

        # Define columns related to different platforms
    platform_columns = [
        'Spotify Streams', 'YouTube Views', 'TikTok Views',
        'Apple Music Playlist Count', 'Amazon Playlist Count',
        'Pandora Streams', 'Soundcloud Streams'
    ]

    # Dictionary to store top 3 songs for each platform
    top_songs_per_platform = {}

    # For each platform, get the top 3 unique songs based on the highest values in each column
    for platform in platform_columns:
        # Convert the column to numeric in case of non-numeric values
        workingFrame[platform] = pd.to_numeric(workingFrame[platform].replace({',': ''}, regex=True), errors='coerce')

        # Get top 3 unique songs by dropping duplicates in the 'Track' column
        top_songs_per_platform[platform] = workingFrame.drop_duplicates(subset='Track').nlargest(3, platform)[['Track', platform]]

    # Plotting the top 3 songs for each platform
    fig, axes = plt.subplots(len(platform_columns), 1, figsize=(10, 18), constrained_layout=True)

    for i, platform in enumerate(platform_columns):
        top_songs = top_songs_per_platform[platform]
        axes[i].barh(top_songs['Track'], top_songs[platform], color='skyblue')
        axes[i].set_title(f"Top 3 Songs on {platform}")
        axes[i].set_xlabel('Stream/View Count')
        axes[i].invert_yaxis()  # Highest value at the top

    plt.tight_layout()
    plt.show()




if __name__ == "__main__":
    workingFrame = dataImport("/content/Most Streamed Spotify Songs 2024 - UTF-8.csv")
    streamFrame = createDataSubset(workingFrame)
    timCode()
    sakeetCode()
    umaCode()
    anilCode()

```
